```python
import os
import glob
import pandas as pd

path = os.path.join('data','reporte-proyecto-ley-20231006055139.xlsx')
distritos = pd.read_excel(path, header=5)
```

```python
distritos.columns
```

```python
columnNames=['Proyecto de Ley', 'Fecha de Presentación',
       'Título', 'Estado', 'Proponente', 'Autores']
distritos=distritos.loc[:,columnNames]
```

```python
#filtro
palabras_deseadas = ['creación del distritos', 'creación del distrito', 'Ley que declara de interés nacional y necesidad pública la creación del distrito']
cre_distritos=distritos[distritos['Título'].str.contains('|'.join(palabras_deseadas), case=False)]
cre_distritos
```

```python
#limpieza
import re
cre_distritos['provincia']=cre_distritos['Título'].str.extract(r'provincia de (.*?) (?:departamento|región)', flags=re.IGNORECASE)
cre_distritos['provincia'] = cre_distritos['provincia'].str.replace(r'\s+(?:DEL|DE)?,?$', '', regex=True, flags=re.IGNORECASE)
```

```python
cre_distritos['provincia'] = cre_distritos['provincia'].str.replace(r'[.,"]','', regex=True)
cre_distritos['provincia'] = cre_distritos['provincia'].str.strip().str.title()
```

```python
cre_distritos['departamento'] = cre_distritos['Título'].str.extract(r'(?:departamento|región) (.*?)$', flags=re.IGNORECASE)
```

```python
cre_distritos['departamento']=cre_distritos['departamento'].str.replace(r'\bDE\s+', '', regex=True, flags=re.IGNORECASE). str.replace (r'[.,"]','', regex=True)
cre_distritos['departamento']=cre_distritos['departamento'].str.strip().str.title()
cre_distritos
```

```python
cre_distritos['distritos_propuesto'] = cre_distritos['Título'].str.extract(r'creación del distrito (.*?) provincia', flags=re.IGNORECASE)
cre_distritos
```

```python
cre_distritos['distritos_propuesto']=cre_distritos['distritos_propuesto'].str.strip().str.replace(r'^(en|la|de|de la|en la)\s+|(en|la|de la|en la)\s+$', '', regex=True, flags=re.IGNORECASE).str.replace(r'\s+EN\s+LA$', '', regex=True, flags=re.IGNORECASE).str.replace(r'[.,"]','', regex=True)
cre_distritos['distritos_propuesto']=cre_distritos['distritos_propuesto'].str.title()
```

```python
import unicodedata

def eliminar_acentos(cadena):
    return ''.join((c for c in unicodedata.normalize('NFD', str(cadena)) if unicodedata.category(c) != 'Mn'))

# Lista de nombres de columnas en las que se eliminarán acentos
columnas_acentos = ['provincia', 'departamento', 'distritos_propuesto']

# Aplicar la función solo a las columnas seleccionadas
for columna in columnas_acentos:
    cre_distritos[columna] = cre_distritos[columna].apply(eliminar_acentos)
```

```python
import os
cre_distritos.to_excel(os.path.join("data", "distritos_2021-2022_limpieza1.xlsx"))
```
