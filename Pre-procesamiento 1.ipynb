```python
import os
import pandas as pd

path = os.path.join('data','distritos_2011-2023.xlsx')
distritos = pd.read_excel(path)
```

```python
distritos=distritos.set_index('Proyecto de Ley')
```

```python
columnNames=['provincia', 'departamento', 'distrito_propuesto']
distritos_propuesto=distritos.loc[:,columnNames]
distritos_propuesto
```

```python
distritos_propuesto['distrito_propuesto'] = distritos_propuesto['distrito_propuesto'].str.split(' / ')
distritos_propuesto = distritos_propuesto.explode('distrito_propuesto')
distritos_propuesto
```

```python
distritos_propuesto['distrito_propuesto'] = distritos_propuesto['distrito_propuesto'].str.split('&')
distritos_propuesto = distritos_propuesto.explode('distrito_propuesto')
distritos_propuesto
```

```python
distritos_propuesto.to_excel(os.path.join("data", "distritos_propuesto.xlsx"))
```
