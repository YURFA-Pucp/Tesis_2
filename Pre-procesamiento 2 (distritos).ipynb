```python
import os
import pandas as pd

path = os.path.join('data','distritos_propuesto_tiempo.xlsx')
distritos = pd.read_excel(path)
distritos
```

```python
distritos=distritos.set_index('Proyecto de Ley')
duplicados = distritos.duplicated(keep=False)
f_duplicadas = distritos[duplicados]
distritos=distritos.drop_duplicates()
distritos
```

```python
#Agregar o quitar columna de acuerdo sea el caso. Por ejemplo cuando se agrega proponente, lo agregamos en el coÃ³digo que sigue:

duplicados = distritos[distritos.duplicated(subset=["provincia", "departamento", "distrito_propuesto", "Periodo"], keep=False)]

# Filtra las filas duplicadas para quedarte con aquellas que tienen al menos un "SÃ­" en la columna "Creado"
duplicados_filtrados = duplicados[duplicados["Creado"] == "Si"]

# Combina el DataFrame original con las filas no duplicadas y las filas duplicadas filtradas
distritos = pd.concat([distritos[~distritos.index.isin(duplicados.index)], duplicados_filtrados])
distritos
```

```python
distritos.to_excel(os.path.join("data", "distritos_propuesto_tiempo_SD.xlsx"))
```
