```python
import os
import pandas as pd

path = os.path.join('data','distritos_2021-2022.xlsx')
distritos_2021_2023 = pd.read_excel(path)

path2 = os.path.join ('data', 'BD_PL_Creacion_disitritos_2011-2022.xlsx')
distritos_2011_2021 = pd.read_excel(path2)
```

```python
#Agregando periodo
periodo = '2021-2023'
distritos_2021_2023['Periodo'] = periodo
#
```

```python
def condicional_1(row):
    if row['Proponente'] == 'Poder Ejecutivo':
        return 'Ley de creación de distrito'
    else:
        return 'Declaratoria de interés'
    
distritos_2021_2023['Tipo'] = distritos_2021_2023.apply(condicional_1, axis=1)
```

```python
#Union de los distritos
distritos_2011_2023= pd.concat([distritos_2021_2023, distritos_2011_2021], axis=0, ignore_index=True)
```

```python
columnNames=['Proyecto de Ley', 'Fecha de Presentacion',
       'Titulo', 'Estado', 'Proponente', 'Autores', 'provincia', 'departamento', 'distrito_propuesto', 'Periodo', 'Tipo']
distritos_2011_2023=distritos_2011_2023.loc[:,columnNames]
```

```python
distritos_2011_2023['Estado']=distritos_2011_2023['Estado'].replace({'Publicada en el Diario Oficial El Peruano': 'Publicado El Peruano', 'Retirado por su Autor': 'Retirado'})
distritos_2011_2023['Proponente']=distritos_2011_2023['Proponente'].replace({'Poder Ejecutivo': 'Ejecutivo'})
```

```python
def condicional_2(row):
    if row['Proponente'] == 'Ejecutivo' and row['Estado'] == 'Publicado El Peruano':
        return 'Si'
    else:
        return 'No'
    
distritos_2011_2023['Creado'] = distritos_2011_2023.apply(condicional_2, axis=1)
```

```python
distritos_2011_2023= distritos_2011_2023.set_index('Proyecto de Ley')
```

```python
import unicodedata

def eliminar_acentos(cadena):
    return ''.join((c for c in unicodedata.normalize('NFD', str(cadena)) if unicodedata.category(c) != 'Mn'))

# Lista de nombres de columnas en las que se eliminarán acentos
columnas_acentos = ['provincia', 'departamento', 'distrito_propuesto']

# Aplicar la función solo a las columnas seleccionadas
for columna in columnas_acentos:
    distritos_2011_2023[columna] = distritos_2011_2023[columna].apply(eliminar_acentos)
```
